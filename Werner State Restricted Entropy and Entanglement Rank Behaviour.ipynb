# Werner State Restricted Entropy and Entanglement Rank Behaviour

This notebook generates the non-multiplicativity of the entanglement rank cones and the non-additivity of the entanglement rank entropy of the Werner states: 
$$\rho_{\lambda} = \frac{\lambda}{\binom{n+1}{2}}\Pi_{+} + \frac{1-\lambda}{\binom{n}{2}}\Pi_{-} \ , $$
where $\lambda \in [0,1]$ and $\Pi_+ , \Pi_-$ are the symmetric and anti-symmetric projectors respectively.

```python
import matplotlib.pyplot as plt
import numpy as np 
import os
file_base = os.getcwd()
```

## Schmidt Norm Results

```python
file_to_read = os.path.abspath(file_base+"/norm_results.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,7)) #This gets rid of info column
```

```python
x_axis = data[1:,1]
#Plot Lines
plt.plot(x_axis,data[1:,2], label = "$\|\| \\rho_{\lambda}^{\otimes 2} \|\|_{S(1)}$")
plt.plot(x_axis,data[1:,3], label = "$\|\| \\rho_{\lambda} \|\|_{S(1)}^{2}$")
plt.plot(x_axis,data[1:,4], label = "$\|\| \\rho_{\lambda}\|\|_{S(k \geq 2)}^{2}$")
plt.axvline(x=0.3, color='purple', linestyle='--')
plt.axvline(x=0.6666, color='purple', linestyle='--')

#plt.xlim([0,1])
#plt.ylim([0,12.5])

plt.xlabel('State Parameter, $\lambda$')
plt.title('Operator Schmidt Norm $S(k)$ for $d=3$ Werner States')
plt.legend()
plt.figure(dpi=1500)
plt.show()
```

```python
x_axis = data[1:,1]
#Plot Lines
plt.plot(x_axis,data[1:,5], label = "$\|\| \\rho_{\lambda}^{\otimes 2} \|\|_{S(1)} - \|\| \\rho_{\lambda}\|\|_{S(1)}^{2} $")
plt.plot(x_axis,data[1:,6], label = "$\|\| \\rho_{\lambda}^{\otimes 2} \|\|_{S(1)} - \|\| \\rho_{\lambda}\|\|_{S(k \geq 2)}^{2}$")
plt.axvline(x=0.6666, color='purple', linestyle='--')
plt.axvline(x=0.3, color='purple', linestyle='--')

#plt.xlim([0,1])
plt.ylim([-0.025,0.025])

plt.xlabel('State Parameter, $\lambda$')
plt.title('Non-Multiplicativity of $S(k)$ for $d=3$ Werner States')
plt.legend()
plt.figure(dpi=1500)
plt.show()
```

## Entanglement Rank Restricted Entropy Results

```python
file_to_read = os.path.abspath(file_base+"/entropy_results.csv")
data = np.genfromtxt(file_to_read,delimiter=',',usecols=np.arange(0,7)) #This gets rid of info column
```

```python
x_axis = data[1:,1]
#Plot Lines
plt.plot(x_axis,data[1:,2], label = "$H_{\min}^{Ent_{1}}(A^{2}_{1}|B^{2}_{1})_{\\rho_{\lambda}^{\otimes 2}}$")
plt.plot(x_axis,data[1:,3],'--', markersize=3, label = "$2 H_{\min}^{Ent_{1}}(A|B)_{\\rho_{\lambda}}$")
plt.plot(x_axis,data[1:,4],'--', markersize=3, label = "$2 H_{\min}^{Ent_{k \geq 2}}(A|B)_{\\rho_{\lambda}}$")
plt.axvline(x=0.6666, color='purple', linestyle='--')

plt.xlabel('State Parameter, $\lambda$')
plt.title('Entanglement-Rank Restricted Min-Entropy of Werner States')
plt.legend()
plt.figure(dpi=1500)
plt.show()
```

```python
x_axis = data[1:,1]
#Plot Lines
plt.plot(x_axis,data[1:,5], label = "$H_{\min}^{Ent_{1}}(A^{2}_{1}|B^{2}_{1})_{\\rho_{\lambda}^{\otimes 2}} - 2H^{Ent_{1}}_{\min}(A|B)_{\\rho_{\lambda}}$")
plt.plot(x_axis,data[1:,6], label = "$H_{\min}^{Ent_{1}}(A^{2}_{1}|B^{2}_{1})_{\\rho_{\lambda}^{\otimes 2}} - 2H^{Ent_{k \geq 2}}_{\min}(A|B)_{\\rho_{\lambda}}$")
plt.axvline(x=0.6666, color='purple', linestyle='--')

plt.xlabel('State Parameter, $\lambda$')
plt.title('Non-Additivity of $H_{\min}^{Ent_{1}}$ for Werner States')
plt.legend()
plt.figure(dpi=1500)
plt.show()
```

```python
x_axis = data[1:,1]
#Plot Lines
plt.plot(x_axis,data[1:,5], label = "$H_{\min}^{Ent_{1}}(A^{2}_{1}|B^{2}_{1})_{\\rho_{\lambda}^{\otimes 2}} - 2H^{Ent_{1}}_{\min}(A|B)_{\\rho_{\lambda}}$")
plt.plot(x_axis,data[1:,6], label = "$H_{\min}^{Ent_{1}}(A^{2}_{1}|B^{2}_{1})_{\\rho_{\lambda}^{\otimes 2}} - 2H^{Ent_{k \geq 2}}_{\min}(A|B)_{\\rho_{\lambda}}$")
plt.axvline(x=0.6666, color='purple', linestyle='--')

plt.xlabel('State Parameter, $\lambda$')
plt.title('Non-Additivity of $H_{\min}^{Ent_{1}}$ for Werner States')
plt.legend()
plt.figure(dpi=1500)
plt.show()
```
